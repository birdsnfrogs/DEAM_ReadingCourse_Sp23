# An example of a Fourier Series

Consider the function $f:\mathbb{R}\to\mathbb{R}$ which is $2\pi$-periodic, and is defined in $[-\pi,\pi]$ by the formula

$$ f(x) = \left \{ \begin{array}{l} 1 \text{ if } |x|\leq 1 \\ 0 \text{ if } |x|>1  \end{array} \right.$$

Let us compute its Fourier series,

$$ f \sim \tfrac{1}{2}a_0 + \sum \limits_{n=1}^\infty a_n \cos(nx) + b_n\sin(nx)  $$

Here, we recall that for $n \in \mathbb{N}$

$$ a_n = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x)\cos(nx)\;dx,\; b_n = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x)\sin(nx)\;dx$$

and,

$$ a_0 = \frac{1}{\pi}\int_{-\pi}^{\pi} f(x)\;dx$$

**Note:** The function $f(x)$ is even, namely $f(x)=f(-x)$, from here one can see that $b_n=0$ for every $n$ (Why is this?).

A direct computation shows that for the concrete function $f$, we have

$$a_0 = \frac{2}{\pi},\; a_n = \frac{2\sin(n)}{\pi n} \text{ for } n \in \mathbb{N}$$

We are going to use pyplot to visualize the partial sum $f_n$, where

$$ f_N(x) = \frac{1}{2}a_0 + \sum \limits_{n=1}^N a_n \cos(nx)$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set a grid for the interval [-\pi,\pi]

K = 1000 # Number of grid points
N = 200 # Length of the Fourier Series
x = np.linspace(-np.pi,np.pi,K)
a = np.empty(N+1)
fn = np.empty((K,len(x)))  # Finite Trigonometric Series of f
```

```python
a[0] = 2/np.pi

for n in range(1,N+1):
    a[n] = 2*np.sin(n)/((np.pi)*n)
    
for n in range(N):
    fn[n] = (0.5)*a[0] 
    for k in range(1,n):
        fn[n] = fn[n] + a[k]*np.cos(k*x)
```

```python
plt.plot(x,fn[5],color='purple')
plt.plot(x,fn[10],color='blue')
plt.plot(x,fn[199],color='red')
```
